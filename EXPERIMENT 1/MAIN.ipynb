```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/kaggle/input/ecommerce-data/data.csv", encoding="ISO-8859-1")

print("Harshavarthan SB 24BAD033/n")
print(data.head())
print(data.tail())
print(data.info())
print(data.describe())
print(data.isnull().sum())
data = data.dropna()


data["Sales"] = data["Quantity"] * data["UnitPrice"]

product_sales = data.groupby("Description")["Sales"].sum().head(10)

plt.figure()
product_sales.plot(kind="bar")
plt.title("Sales per Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()

plt.figure()
product_sales.plot(kind="line")
plt.title("Sales Trend per Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/kaggle/input/pima-indians-diabetes-database/diabetes.csv")

print("Harshavarthan SB 24BAD033/n")
print(data.head())
print(data.tail())
print(data.info())
print(data.describe())
print(data.isnull().sum())


cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
for c in cols:
    data[c] = data[c].replace(0, pd.NA)

print(data.isnull().sum())

plt.figure()
plt.hist(data["Glucose"].dropna())
plt.title("Glucose Level Distribution")
plt.xlabel("Glucose")
plt.ylabel("Count")
plt.show()

plt.figure()
plt.hist(data["Age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

plt.figure()
plt.boxplot(data["Glucose"].dropna())
plt.title("Glucose Level Boxplot")
plt.ylabel("Glucose")
plt.show()

plt.figure()
plt.boxplot(data["Age"])
plt.title("Age Boxplot")
plt.ylabel("Age")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("/kaggle/input/housing-price-prediction/Housing.csv")

print("Harshavarthan SB 24BAD033/n")
print(data.columns)
print(data.head())
print(data.info())
print(data.isnull().sum())
print(data.isnull().sum()[data.isnull().sum() > 0])

# scatter plot: area vs price
plt.figure()
plt.scatter(data["area"], data["price"])
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()

# scatter plot: bedrooms vs price
plt.figure()
plt.scatter(data["bedrooms"], data["price"])
plt.title("Bedrooms vs Price")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()

# correlation matrix
corr = data.corr(numeric_only=True)

# heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/kaggle/input/customer-personality-analysis/marketing_campaign.csv", sep="\t")

print("Harshavarthan SB 24BAD033/n")
print(data.head())
print(data.info())
print(data.columns)
print(data.isnull().sum())
data = data.dropna()

data["Age"] = 2024 - data["Year_Birth"]

plt.figure()
data["Age"].value_counts().head(10).plot(kind="bar")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# box plot for income
plt.figure()
plt.boxplot(data["Income"])
plt.title("Income Distribution")
plt.ylabel("Income")
plt.show()

# total spending calculation
data["Total_Spending"] = (
    data["MntWines"] +
    data["MntFruits"] +
    data["MntMeatProducts"] +
    data["MntFishProducts"] +
    data["MntSweetProducts"] +
    data["MntGoldProds"])

# box plot for spending
plt.figure()
plt.boxplot(data["Total_Spending"])
plt.title("Customer Spending Pattern")
plt.ylabel("Total Spending")
plt.show()
```
